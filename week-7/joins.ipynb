```python
#create dataframe

data = {'Make': ['Toyota', ' Lexus', ' BMW', 'Hyundai'],
    'Model': ['Rav4', 'LS20', 'X3', 'Elantra'],
    'Year': [2018, 2015, 2012, 2022],
    'Color': ['White', 'Black', ' Blue', ' Green']}

import pandas as pd

cars = pd.DataFrame(data)
cars
```

```python
more_data = {'Make': ['Toyota', ' Lexus', ' BMW', 'Hyundai'],
    'Year': [2019, 2011, 2007, 2052],
    'Model': ['Corolla', 'LS50', 'X36', 'Not Sure'],
    'Color': ['White', 'Black', ' Blue', ' Green']}

more_cars = pd.DataFrame(more_data)
more_cars
```

```python
#Combine both dataset

pd.concat([cars, more_cars], axis=0)
```

```python
price_data = {'Price': [35000, 50000, 54000, 27000], 'Seats': [5, 5, 5, 5]}
prices = pd.DataFrame(price_data)
prices
```

```python
pd.concat([cars, prices], axis=1)
```

```python
movies = pd.read_csv('movies.csv')
actors = pd.read_csv('actors.csv')
```

```python
#to get the actor data for movies, join the movies and actors dataframes
pd.merge(movies, actors, how='inner', on='movie_id')
```

```python
pd.merge(movies, actors, how='left', on='movie_id')
```
