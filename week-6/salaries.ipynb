```python
#filter base pay greater than 80k
#show employee name and job title only

salaries[salaries['BasePay'] >= 80000][['EmployeeName', 'JobTitle']]
```

```python
#multiple conditions
#basepay greater than 80k from the San Francisco agency

#1 wrap each condition in round brackets
#2. we use '&' instead of the word 'and'

salaries[(salaries['BasePay'] >= 80000) & (salaries['Agency'] >= "San Francisco")][['BasePay', 'EmployeeName']]
```
